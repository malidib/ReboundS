```python
import rebound
import reboundx

# Simulation settings
TIMESTEP_YR = 1.0e4
INTEGRATION_SPAN_YR = 1.0e9
EDDINGTON_WIND_CONST = 1.0e-6

# Create the simulation and set units
sim = rebound.Simulation()
sim.units = ("AU", "yr", "Msun")
sim.integrator = "whfast"
sim.dt = TIMESTEP_YR

# Add a single massive star
sim.add(m=100.0)
star = sim.particles[0]

# Attach REBOUNDx
rebx = reboundx.Extras(sim)

# Add stellar evolution first so stellar properties are initialized
sse = rebx.load_operator("stellar_evolution_sse")
rebx.add_operator(sse)
sim.step()

# Add Eddington winds after SSE is active
edw = rebx.load_operator("eddington_winds")
rebx.add_operator(edw)
edw.params["edw_const"] = EDDINGTON_WIND_CONST

# Print initial stellar state
print(star.m, star.r, star.params["sse_L"])

# Evolve forward in time
sim.integrate(sim.t + INTEGRATION_SPAN_YR)

# Print final stellar state
print(star.m, star.r, star.params["sse_L"])
```
